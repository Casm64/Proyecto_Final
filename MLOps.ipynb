```python

df=pd.read_csv("movies_dataset.csv")

#Función para desanidar columnas JSON
def desanidar_columna(df,columna):
    if columna in df.columns:
        desanidado=pd.json_normalize(df[columna].dropna().apply(eval))
        desanidado.columns=[f"{columna}_{subcol}"for subcol in desanidado.columns]
        df=df.drop(columns=[columna]).join(desanidado,how='left')
    return df

#Columnas para desanidar
columnas_anidadas=[
    'belongs_to_collection',
    'production_companies',
    'genres',
    'production_countries',
    'spoken_languages'
]

#Para desanidar cada columna hice un ciclo for
for col in columnas_anidadas:
    df=desanidar_columna(df,col)

#Guardar el DataFrame desanidado en un nuevo archivo Excel
archivo_nuevo='movies_dataset_des.xlsx'
df.to_excel(archivo_nuevo, index=False)

print("Archivo de excel guardado satisfactoriamente", archivo_nuevo)
```
